```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Parameters
num_samples = 72  # set your desired number of samples
mean = 33/1000           # set your desired mean
std_dev = 3/1000        # set your desired standard deviation

# Sampling
samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# Reporting min and max
min_val = np.min(samples)
max_val = np.max(samples)
print(f"Min value: {min_val}")
print(f"Max value: {max_val}")

# Plotting histogram
plt.hist(samples, bins=30, edgecolor='black')
plt.title('Histogram of Gaussian Samples')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
print(f"Seam Size: {int(round(min_val * 1000, 0))}/1000 to {int(round(max_val * 1000, 0))}/1000 in")
```

```python
# Generate 100 samples of the sum of 5 independent uniform[0,1] variables
num_trials = 100000
x = np.random.uniform(0, 1, size=(num_trials, 5))
sums = np.sum(x, axis=1)

# Plot histogram
plt.hist(sums, bins=20, edgecolor='black')
plt.title('Histogram of Sum of 5 Uniform[0,1] Variables (100,000 Trials)')
plt.xlabel('Sum')
plt.ylabel('Frequency')
plt.gca().yaxis.set_major_locator(MultipleLocator(5000))  # Major ticks every 5
plt.gca().yaxis.set_minor_locator(MultipleLocator(2500))  # Minor ticks every 2.5
plt.show()
```

```python
# Probability mass function for the sum of two dice
sums_two_dice = np.arange(2, 13)
pmf = np.array([1,2,3,4,5,6,5,4,3,2,1]) / 36

plt.stem(sums_two_dice, pmf, basefmt=" ")
plt.title('Probability Mass Function of the Sum of Two Dice')
plt.xlabel('Sum')
plt.ylabel('Probability')
plt.xticks(sums_two_dice)
plt.show()
```

```python
# Sample 1000 random values of x uniformly from [0, 1)
x_sin = np.random.uniform(0, 1, 100000)
y_sin = np.sin(2 * np.pi * x_sin)

# Plot histogram of y values
plt.hist(y_sin, bins=30, edgecolor='black', density=False)
plt.title('Histogram of y = sin(2x)')
plt.xlabel('X')
plt.ylabel('f(x)')
#plt.gca().yaxis.set_major_locator(MultipleLocator(0.5))  # Major ticks every 0.5
#plt.gca().yaxis.set_minor_locator(MultipleLocator(0.25))  # Minor ticks every 0.25
plt.gca().xaxis.set_major_locator(MultipleLocator(0.5))  # Major ticks every 0.5
plt.gca().xaxis.set_minor_locator(MultipleLocator(0.25))  # Minor ticks every 0.25
plt.show()
```
