```python
import numpy as np
import math
```

```python
def velocity_statistics(filename):
    data = np.loadtxt(filename, skiprows=1)

    N = data.size
    if N == 0:
        raise ValueError("No data points found after skipping the first line.")
    else:
        mean = np.mean(data)
        std = np.std(data, ddof=1)
        std_of_mean = std / math.sqrt(N)
        print(f"N={N}, V̄={mean:.1f} m/s, s_V={std:.2f} m/s, s_V̄={std_of_mean:.2f} m/s")
        print(f"V̄ = {mean:.1f} ± {std_of_mean * 1.96:.2f} m/s (95% confidence interval)\n")

    return {
        "count": N,
        "mean": mean,
        "std_dev": std,
        "sem": std_of_mean
    }
```

```python
print("BatSpeedSS:")
ss = velocity_statistics("BatSpeedSS.txt")
```
